```python
import pandas as pd
import numpy as np

data = pd.read_csv('OnlineRetail.csv')
data
```

```python
# Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.
row = data.shape[0]
col = data.shape[1]
print(f'Số dòng là {row}, số cột là {col}')
print('Kiểu dữ liệu của các thuộc tính:\n',data.dtypes)
```

```python
# Xây dựng bảng Pivot table, với mỗi Số hóa đơn tính trung bình cộng số lượng các mặt hàng theo từng Quốc gia. 
quantity_mean = pd.pivot_table(data,values='Quantity',index=['InvoiceNo'],columns=['Country'],aggfunc=np.mean)
quantity_mean
```

```python
# Xây dựng bảng Pivot table, với mỗi Khách hàng cho biết số lượng mua hàng lớn nhất và nhỏ nhất theo Kho.
quantity_min_max = pd.pivot_table(data,values='Quantity',index=['CustomerID'],columns=['StockCode'],aggfunc={'min','max'})
quantity_min_max
```

```python
# Xây dựng bảng Pivot table, với mỗi Mã kho tính tổng số lượng các mặt hàng và trung bình cộng giá.
quantity_sum_mean = pd.pivot_table(data,values='Quantity',index='StockCode',aggfunc={'sum','mean'})
quantity_sum_mean
```

```python
# Xây dựng bảng Pivot table cho biết tổng số lượng hàng bán được của mỗi ngày.
sum_quantity_by_date = pd.pivot_table(data,values='Quantity',index='InvoiceDate',aggfunc='sum')
sum_quantity_by_date
```
